**Remover coluna Unnamed**

```python
df_2020 = df_2020.loc[:, ~df_2020.columns.str.contains("Unnamed")]
df_2020.columns
```

**Criar coluna de data + hora**

```python
# A coluna "Hora UTC" do INMET vem com valores como "0000 UTC", "0100 UTC", etc.
# Para facilitar a construção de uma data e hora completas, convertemos essa coluna para texto
# e pegamos apenas os dois primeiros caracteres, que representam a hora (00, 01, 02, ...).

# df_2020["Hora UTC"] = df_2020["Hora UTC"].astype(str).str[:2]


# Em seguida, criamos uma nova coluna chamada "datetime".
# Para isso, juntamos a coluna "Data" com a coluna "Hora UTC",
# formando strings como "2020/01/01 00", "2020/01/01 01", etc.
# A função pd.to_datetime() converte essas strings para o tipo datetime do pandas,
# permitindo análises temporais (agrupamento por dia, mês, gráficos de séries temporais, etc.).

'''df_2020["datetime"] = pd.to_datetime(
    df_2020["Data"] + " " + df_2020["Hora UTC"],
    format="%Y/%m/%d %H"
)'''

df_2020["Hora UTC"] = df_2020["Hora UTC"].astype(str).str[:2]

df_2020["datetime"] = pd.to_datetime(
    df_2020["Data"] + " " + df_2020["Hora UTC"],
    format="%Y/%m/%d %H"
)
```

```python
df_2020[["Data","Hora UTC","datetime"]].head()
```

**Alterando nome da coluna para algo mais simples**

```python
df_2020 = df_2020.rename(columns={
    "TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)": "temperatura"
})

df_2020
```

**Padronizando a coluna temperatura como texto, trocando vírgulas por pontos depois convertendo texto em número**

Obs: O errors="coerce" quer dizer:
se aparecer algum valor estranho que não dá para converter, o pandas transforma em NaN em vez de quebrar o código.

```python
df_2020["temperatura"] = df_2020["temperatura"].astype(str)
df_2020["temperatura"] = df_2020["temperatura"].str.replace(",", ".", regex=False)
df_2020["temperatura"] = pd.to_numeric(df_2020["temperatura"], errors="coerce")
```

**Conferindo se funcionou**

```python
df_2020["temperatura"].dtype
```

**Agrupando os dados por dia, pegando a coluna temperatura e calculando a média de cada dia. Depois, checando se funcionou.**

Obs: .dt é um acessador de propriedades de data e hora do pandas. Ele permite extrair partes da data.

```python
temp_diaria = df_2020.groupby(df_2020["datetime"].dt.date)["temperatura"].mean()
temp_diaria.head()
```

A variável "temp_mensal" contém a média da temperatura para cada mês do ano.

**Leitura, tratamento e concatenação automática dos arquivos do INMET.**

Este bloco percorre todos os arquivos CSV da estação meteorológica presentes no diretório e aplica o mesmo processo de preparação dos dados em cada um deles.

As etapas incluem:
- leitura dos arquivos CSV com os parâmetros corretos de separador e encoding;
- remoção de colunas desnecessárias (como colunas "Unnamed");
- limpeza da coluna de hora (mantendo apenas a hora numérica);
- criação de uma coluna datetime combinando data e hora;
- renomeação da coluna de temperatura para um nome mais simples;
- conversão da temperatura para formato numérico.

Cada dataframe tratado é armazenado em uma lista e, ao final do processo, todos os dataframes são concatenados em um único dataframe chamado df_total, que reúne os dados de todos os anos disponíveis.

**Ler e tratar cada arquivo**

Agora vamos aplicar o mesmo tratamento que usamos para 2020 em todos os arquivos.

```python
import os

arquivos = [f for f in os.listdir() if f.startswith("INMET") and f.endswith(".CSV")]

sorted(arquivos)

lista_dfs = []
```

```python
import pandas as pd

for arquivo in arquivos:

    df = pd.read_csv(
        arquivo,
        sep=";",
        encoding="latin1",
        skiprows=8
    )

    # remover colunas Unnamed
    df = df.loc[:, ~df.columns.str.contains("Unnamed", na=False)]

    # limpar hora
    df["Hora UTC"] = df["Hora UTC"].astype(str).str[:2]

    # criar datetime
    df["datetime"] = pd.to_datetime(
        df["Data"] + " " + df["Hora UTC"],
        format="%Y/%m/%d %H"
    )

    # renomear coluna temperatura
    df = df.rename(columns={
        "TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)": "temperatura"
    })

    # converter temperatura
    df["temperatura"] = df["temperatura"].astype(str)
    df["temperatura"] = df["temperatura"].str.replace(",", ".", regex=False)
    df["temperatura"] = pd.to_numeric(df["temperatura"], errors="coerce")

    # guardar dataframe tratado
    lista_dfs.append(df)
```

Juntar todos os anos em um único dataset.

```python
df_total = pd.concat(lista_dfs, ignore_index=True)

df_total.head(10)
```

**Verificando se deu certo:**

```python
df_total["datetime"].dt.year.value_counts().sort_index()
```

**Cálculo da temperatura média anual.**

Neste passo, os dados de temperatura são agrupados por ano utilizando a coluna "ano", extraída da coluna datetime.

Para cada ano, é calculada a média das temperaturas registradas ao longo de todo o período disponível. Esse tipo de agregação permite observar possíveis variações climáticas entre os anos e identificar tendências de aquecimento ou resfriamento.

```python
df_total["ano"] = df_total["datetime"].dt.year

temp_anual = df_total.groupby("ano")["temperatura"].mean()
temp_anual
```

**Visualizando:**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.plot(temp_anual.index, temp_anual.values, marker="o")

plt.title("Temperatura média anual - São Paulo (Interlagos)")
plt.xlabel("Ano")
plt.ylabel("Temperatura média (°C)")

plt.show()
```

**O que podemos observar**

2021–2022: foram os anos mais frios da série.

Isso pode estar relacionado a fenômenos climáticos como La Niña, que costuma resfriar parte do Brasil.

2023–2024: aqui aparece um aumento claro na temperatura média.

2024 foi o ano mais quente da série: 20.38 °C

Isso é interessante porque 2023–2024 foram anos globalmente muito quentes, segundo relatórios da NOAA e NASA.

2025: A média caiu um pouco novamente.

**Cálculo da tendência linear da temperatura média anual.**

Neste bloco, utilizamos regressão linear simples para ajustar uma reta aos valores de temperatura média anual ao longo do tempo.

A variável "anos" representa o eixo temporal, enquanto "temperaturas" contém as médias anuais calculadas anteriormente.
A função linregress() estima os parâmetros da reta:
- slope: inclinação da reta, indicando a variação média da temperatura por ano;
- intercept: intercepto da reta.

Em seguida, calculamos os valores da linha de tendência com a equação: temperatura_estimada = intercept + slope * ano

O gráfico final compara os valores observados com a tendência linear estimada, permitindo visualizar se há um padrão geral de aumento ou diminuição da temperatura no período analisado.

```python
from scipy.stats import linregress

anos = temp_anual.index
temperaturas = temp_anual.values

resultado = linregress(anos, temperaturas)

slope = resultado.slope
intercept = resultado.intercept

tendencia = intercept + slope * anos
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

# dados reais
plt.plot(anos, temperaturas, marker="o", label="Temperatura média")

# linha de tendência
plt.plot(anos, tendencia, linestyle="--", label="Tendência linear")

plt.title("Tendência de temperatura média anual - São Paulo (Interlagos)")
plt.xlabel("Ano")
plt.ylabel("Temperatura média (°C)")

plt.legend()

plt.show()
```

**Explicando o gráfico**

A linha azul mostra os valores reais de cada ano, e a linha laranja tracejada mostra a tendência linear estimada. Como ela está subindo, o ajuste indica uma tendência positiva no período analisado.

Mas tem um ponto importante:

isso não prova sozinho um aquecimento climático robusto, porque sua série é curta, com só 6 anos. O que dá para dizer é:

entre 2020 e 2025, os dados da estação de Interlagos mostram uma tendência linear crescente
essa tendência deve ser interpretada com cautela, porque poucos anos podem ser muito influenciados por variabilidade natural
